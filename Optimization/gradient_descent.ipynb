```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data: y = 3x + 4 + noise
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
# Hyperparameters
learning_rate = 0.1
n_iterations = 1000
m = len(X)  # Number of samples
```

```python
# Initialize weights randomly (theta = [bias, slope])
theta = np.random.randn(2, 1)
```

```python
# Add bias term (x0 = 1) to X
X_b = np.c_[np.ones((m, 1)), X]  # Adds a column of ones to X
```

```python
# Gradient Descent Algorithm
loss_history = []  # Store loss for visualization
```

```python
for iteration in range(n_iterations):
    gradients = (2/m) * X_b.T.dot(X_b.dot(theta) - y)
    theta -= learning_rate * gradients
    
    # Compute loss (Mean Squared Error)
    loss = (1/m) * np.sum((X_b.dot(theta) - y) ** 2)
    loss_history.append(loss)

# Final optimized weights
print(f"Optimized theta (bias, slope): \n{theta}")
```

```python
# Plot the loss curve
plt.figure(figsize=(8, 5))
plt.plot(range(n_iterations), loss_history, label="Loss Curve")
plt.xlabel("Iterations")
plt.ylabel("Mean Squared Error (MSE)")
plt.title("Gradient Descent Convergence")
plt.legend()
plt.show()
```

```python
# Plot the regression line
plt.figure(figsize=(8, 5))
plt.scatter(X, y, label="Data")
plt.plot(X, X_b.dot(theta), color='red', label="Best Fit Line")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression using Gradient Descent")
plt.legend()
plt.show()
```
