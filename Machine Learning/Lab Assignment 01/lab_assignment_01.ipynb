# Problem 3
How many rows are have at least one NA / NULL value? What percentage of the dataset has missing
values?

```python
# Cells
print(sum(data.isnull().values.ravel()))

#Rows
rows = data.isnull().any(axis=1).sum()
print(rows)

total_rows = len(data)
percentage = rows/total_rows * 100
print(f"percentage of rows with missing values: {percentage:.2f}%")
```

```python
data.isnull().sum()
```
