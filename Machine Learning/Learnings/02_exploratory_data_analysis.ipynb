```python
#Histogram
# sns.set_style('whitegrid')
df_USAhousing['median_house_value'].hist(bins=30)
plt.xlabel('median house value')
```

```python
#ScatterPlot
x = df_USAhousing['median_income']
y = df_USAhousing['median_house_value']

plt.scatter(x,y)
plt.show()
```

```python
# Create a jointplot showing "median_income" versus "median_house_value".
# plt.figure(figsize=([10,6]))
sns.jointplot(x='median_income',y='median_house_value',data=df_USAhousing)
# plt.title('Median Income vs Median House Value')
```

```python
# CountPlot
sns.countplot(x='ocean_proximity',data=df_USAhousing)
```

## BigQuery dataset on NYC Taxi Cab rides

```python
# Import
#in terminal pip install google-cloud-bigquery
import os
from google.cloud import bigquery

os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = 'bigquery_ij_service.json'

client = bigquery.Client

sql_query = """
SELECT station_id, name, dockcount
FROM `bigquery-public-data.san_francisco.bikeshare_stations`
LIMIT 50
"""

query_job = client.query(sql_query)

print(query_job)
print(query_job.to_dataframe())

query_job.result()
```
